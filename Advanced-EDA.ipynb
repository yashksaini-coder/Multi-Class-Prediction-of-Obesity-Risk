```python
import sweetviz as sv
# Importing the basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("./playground-series-s4e2/train.csv")
```

```python
df.head()
```

```python

Obesity_report = sv.analyze(df)
Obesity_report.show_html("Obesity-report.html",layout="widescreen",scale=1.0)
```

```python
test_df=pd.read_csv("./playground-series-s4e2/test.csv")
test_df.head()
```

```python
test_train_report = sv.compare([df, "Training Data"], [test_df, "Test Data"])
test_train_report.show_html("Test-Train-report.html",layout="widescreen",scale=1.0)
```
